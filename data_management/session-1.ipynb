```julia
using CSV
using DataFrames
using CategoricalArrays
using FreqTables
```

```julia
# lecture des donnÃ©es complÃ¨te dans un DataFrame
df = DataFrame(CSV.File("bank-additional-full.csv"))
```

```julia
# affichage des 10 premiÃ¨res lignes du fichier
println("Les 10 premiÃ¨res lignes:")
println(first(df, 10))
```

```julia
# 3 derniÃ¨res lignes
println("Les 3 derniÃ¨res lignes:")
println(last(df, 3))
```

```julia
# affichage de la taille du df
println("Taille de la base de donnÃ©es:")
println(size(df))
```

```julia
# filtrer les donnÃ©es pour ne garder que les personnes de plus de 50 ans
df_over_50 = filter(row -> row.age > 50, df)
```

```julia
# que les techniciens de moins de 40 ans
df_tech_under_40 = filter(row -> row.job == "technician" && row.age < 40, df)
```

```julia
# que les personnes entre 30 et 50 ans qui sont propriÃ©taires de leur maison
df_30_to_50_homeowners = filter(row -> row.age >= 30 && row.age <= 50 && row.housing == "yes", df)
```

```julia
# croiser les donnÃ©es entre le statut de propriÃ©taire et le statut d'emprunt
housing_loan_freqtable = freqtable(df, :housing, :loan)
```

```julia
# croiser les donnÃ©es pour mettre en lumiÃ¨re combien de clients existe-t-il par catÃ©gorie professionnelle et statut marital
job_marital_freqtable = freqtable(df, :job, :marital)
```

```julia
# Autre possible croisements d'informations :
# Ã©ducation par tranche d'age

# DÃ©finir les intervalles pour les catÃ©gories d'Ã¢ge
age_bins = 15:15:110 # de 15 Ã  110 ans par tranches de 15 ans

# Utiliser la fonction cut pour crÃ©er une reprÃ©sentation catÃ©gorielle de l'Ã¢ge
df.age_category = cut(df.age, age_bins)

# RÃ©aliser une table de frÃ©quence en fonction de l'Ã©ducation et de la catÃ©gorie d'Ã¢ge
education_age_freqtable = freqtable(df, :education, :age_category)
```
