```julia
using CSV
using DataFrames
using CategoricalArrays
using FreqTables
```

```julia
# lecture des données complète dans un DataFrame
df = DataFrame(CSV.File("bank-additional-full.csv"))
```

```julia
# affichage des 10 premières lignes du fichier
println("Les 10 premières lignes:")
println(first(df, 10))
```

```julia
# 3 dernières lignes
println("Les 3 dernières lignes:")
println(last(df, 3))
```

```julia
# affichage de la taille du df
println("Taille de la base de données:")
println(size(df))
```

```julia
# filtrer les données pour ne garder que les personnes de plus de 50 ans
df_over_50 = filter(row -> row.age > 50, df)
```

```julia
# que les techniciens de moins de 40 ans
df_tech_under_40 = filter(row -> row.job == "technician" && row.age < 40, df)
```

```julia
# que les personnes entre 30 et 50 ans qui sont propriétaires de leur maison
df_30_to_50_homeowners = filter(row -> row.age >= 30 && row.age <= 50 && row.housing == "yes", df)
```

```julia
# croiser les données entre le statut de propriétaire et le statut d'emprunt
housing_loan_freqtable = freqtable(df, :housing, :loan)
```

```julia
# croiser les données pour mettre en lumière combien de clients existe-t-il par catégorie professionnelle et statut marital
job_marital_freqtable = freqtable(df, :job, :marital)
```

```julia
# Autre possible croisements d'informations :
# éducation par tranche d'age

# Définir les intervalles pour les catégories d'âge
age_bins = 15:15:110 # de 15 à 110 ans par tranches de 15 ans

# Utiliser la fonction cut pour créer une représentation catégorielle de l'âge
df.age_category = cut(df.age, age_bins)

# Réaliser une table de fréquence en fonction de l'éducation et de la catégorie d'âge
education_age_freqtable = freqtable(df, :education, :age_category)
```
